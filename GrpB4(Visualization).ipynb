```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\hp\Downloads\airquality.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Ozone"]=df["Ozone"].fillna(df["Ozone"].mean())
df["Solar.R"]=df["Solar.R"].fillna(df["Solar.R"].mean())
df["Humidity"]=df["Humidity"].fillna(df["Humidity"].mode()[0])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df["Humidity"]=label.fit_transform(df["Humidity"])
df["Humidity"].unique()
```

```python
plt.scatter(df["Ozone"],df["Temp"])
plt.title("Ozone Scatter Plot")
plt.legend(["Ozone"])
plt.xlabel("Ozone")
plt.ylabel("Temp")
```

```python
sns.barplot(df["Humidity"],df["Temp"])
```

```python
sns.heatmap(df.corr(),vmin=-1,vmax=1,annot=True)
```

```python
sns.pairplot(df)
```

```python
plt.plot(df.iloc[1:15,0],marker="o",linestyle="dotted",label="Ozone")
plt.plot(df.iloc[1:15,3],marker="o",linestyle="dashed",label="Temp")
plt.title("Line Graph for Ozone")
plt.legend()
plt.show()
```

```python
sns.boxplot(df["Month"],df["Solar.R"])
```

```python
labels=['Ozone','Solar.R','Temp','Wind']
colors=['red','cyan','green','yellow']
sizes=[df["Ozone"].mean(),df["Solar.R"].mean(),df["Temp"].mean(),df["Wind"].mean()]
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
plt.hist(df["Temp"],bins='auto')
plt.title("Histogram")
plt.xlabel("Temp")
```
